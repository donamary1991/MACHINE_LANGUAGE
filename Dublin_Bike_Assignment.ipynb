```python
# Confidence Intervals for two samples

sample1 = df['Bike Rental Cost ($)'].sample(n=50, random_state=42)
sample2 = df['Bike Rental Cost ($)'].sample(n=50, random_state=0)

import scipy.stats as stats

# we are taking 95% confidence level

CI_1 = stats.t.interval(0.95, len(sample1) - 1, loc=np.mean(sample1), scale=stats.sem(sample1))
CI_2 = stats.t.interval(0.95, len(sample2) - 1, loc=np.mean(sample2), scale=stats.sem(sample2))

print(f"Confidence interval for 1st sample is {CI_1}")
print(f"Confidence interval for 2nd sample is {CI_2}")
```

```python
# ANOVA

independents = ["City", "Weather Condition", "Bike Model", "Rider Satisfaction"]

for i in independents:
  print(f"{i} vs Bike Rental Cost")
  print()
  independent_groups = df['Bike Rental Cost ($)'].groupby(df[i])
  f_value, p_value = stats.f_oneway(*[group for name, group in independent_groups])
  print("F-value:", f_value)
  print("P-value:", p_value)

  # critical value

  alpha = 0.05
  dfn = len(df[i].unique()) - 1 # no. of categories - 1
  dfd = len(df) - len(df[i].unique()) # no. of samples - no. of categories
  critical_value = stats.f.ppf(1 - alpha, dfn, dfd)
  print("critical value:",critical_value)
  print()


'''
as per the anova results comparing f value with critical value
we can reject the null hypothesis for Bike Model (since f value>critical value)
but not for the other three, i.e. City and Weather Condition and Rider Satisfaction
'''
```

```python
# t-test

'''
though t tests are implemented when you have 2 categories within a feature,
let's take two categories from the weather condition and find out the t-statistic and
p-value to see if there is a significant difference between them.
'''

weather0 = df[df['Weather Condition'] == 0]['Bike Rental Cost ($)']
weather1 = df[df['Weather Condition'] == 1]['Bike Rental Cost ($)']
t_statistic, p_value = stats.ttest_ind(weather0, weather1)
print("t-statistic:", t_statistic)
print("P-value:", p_value)

'''
since p value> 0.05, we can't reject the null hypothesis
which means that there is no significant difference between the two categories
'''
```

```python
# splitting the dataframe to input and output features
x = df.drop('Bike Rental Cost ($)', axis=1)
y = df['Bike Rental Cost ($)']
```

```python
# splitting the training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Normalisation

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
# Principal Component Analysis

from sklearn.decomposition import PCA
pca = PCA(0.88)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
x_train.shape # features reduced to 5 by PCA
```

```python
# Model Creation and Performance Evaluation using R2_Score and MAPE

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_absolute_percentage_error

linear = LinearRegression()
tree = DecisionTreeRegressor()
forest = RandomForestRegressor()
sv = SVR()
knn = KNeighborsRegressor()

lst=[linear,tree,forest,sv,knn]

for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  r2=r2_score(y_test,y_pred)
  mape=mean_absolute_percentage_error(y_test,y_pred)
  print(f'{i}: \nR2score: {r2} \nMAPE: {mape} \n\n')
  print()
```

```python
# Hyperparameter Tuning

from sklearn.model_selection import GridSearchCV
cls1 = RandomForestRegressor()
param1 = {'n_estimators': [50, 100, 200], 'criterion' : ['squared_error', 'absolute_error'], 'max_depth': [None, 10, 20]}
gv = GridSearchCV(cls1, param1, cv=5, scoring="accuracy")
gv.fit(x_train, y_train)
```

```python
print(gv.best_params_)
```

```python
cls2 = RandomForestRegressor(n_estimators=50)
cls2.fit(x_train, y_train)
y_pred1 =cls2.predict(x_test)
r2=r2_score(y_test,y_pred1)
mape=mean_absolute_percentage_error(y_test,y_pred1)
print(f'R2score: {r2} \nMAPE: {mape}')
```
