```python
df.columns
df.drop(['Unnamed: 0','ID','Type'],axis=1,inplace=True)
```

```python
# Year of Release Analysis:

# Group movies by year and platform, then count
disney_releases = df.groupby('Year')['Disney+'].sum().reset_index()
netflix_releases = df.groupby('Year')['Netflix'].sum().reset_index()
Hulu_releases = df.groupby('Year')['Hulu'].sum().reset_index()
Prime_releases = df.groupby('Year')['Prime Video'].sum().reset_index()

# Create line plot
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
plt.plot(disney_releases['Year'], disney_releases['Disney+'], label='Disney+', marker='o')
plt.plot(netflix_releases['Year'], netflix_releases['Netflix'], label='Netflix', marker='o')
plt.plot(Hulu_releases['Year'], Hulu_releases['Hulu'], label='Hulu', marker='o')
plt.plot(Prime_releases['Year'], Prime_releases['Prime Video'], label='Prime Video', marker='o')


# Set labels and title
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Trend of Movie Releases Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
_#Compare the age restrictions of the movies available on Disney+ and Netflix.
df_Disney=df[df['Disney+']==1]
df_Netflix=df[df['Netflix']==1]

```

```python
# Count movies per age rating for each platform
Disney_age_counts=df_Disney['Age'].value_counts()
Netflix_age_counts=df_Netflix['Age'].value_counts()
Disney_age_counts,Netflix_age_counts
```

```python
# Create a bar plot for comparison
x=np.arange(len(Disney_age_counts))
width=0.35
fig,ax=plt.subplots()
Disney_bar=ax.bar(x-width/2,Disney_age_counts.values,width,label='Disney+')
Netflix_bar=ax.bar(x+width/2,Netflix_age_counts.values,width,label='Netflix')
plt.xlabel('Age Rating')
plt.ylabel('Number of Movies')
plt.title('Comparison of Age Restriction between Disney+ and Netflix')
plt.xticks(x,Disney_age_counts.index)
plt.bar_label(Disney_bar)
plt.bar_label(Netflix_bar)
plt.legend()
plt.show()
```

***Inferences from the Bar Plot:***

* Disney+ has a higher proportion of movies rated for all ages compared to Netflix.

* Netflix has a greater number of movies with higher age restrictions.



```python
# Formatting Rotten Tomatoes scores column
df_Disney['Rotten Tomatoes']=df_Disney['Rotten Tomatoes'].str.replace('/100','').astype(float)
df_Netflix['Rotten Tomatoes']=df_Netflix['Rotten Tomatoes'].str.replace('/100','').astype(float)
df_Disney['Rotten Tomatoes']
```

```python
# Calculate average Rotten Tomatoes scores
avg_Disney_score=df_Disney['Rotten Tomatoes'].mean()
avg_Netflix_score=df_Netflix['Rotten Tomatoes'].mean()
avg_Disney_score,avg_Netflix_score
print(f'Average_Rotten_Tomatoes_Score_for_Disney+:{avg_Disney_score}')
print(f'Average_Rotten_Tomatoes_Score_for_Netflix:{avg_Netflix_score}')
```

```python
# Compare in Rotten Tomatoes Score for movies on Disney+ and Netflix.
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(data=df_Disney['Rotten Tomatoes'])
plt.xlabel('Disney+')
plt.ylabel('Rotten Tomatoes Score')
plt.subplot(1,2,2)
sns.boxplot(data=df_Netflix['Rotten Tomatoes'])
plt.xlabel('Netflix')
plt.ylabel('Rotten Tomatoes Score')
```

***Inferences from the Box Plot:***

* The median line for Disney+ is higher than Netflix, it suggests that Disney+ movies generally have better movies compared to Netflix .
* Netflix has more variability in movie quality compared to Disney+ indicated by high IQR.
