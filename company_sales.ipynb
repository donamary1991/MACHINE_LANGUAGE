```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/company_sales_data.csv')
# df.set_index('month_number',inplace=True)
df
```

```python
# #sales_dataset_matplotlib

# 1. Read Total profit of all months and show it using a line plot
df1=df.groupby('month_number')['total_profit'].sum()
df1
plt.plot(df1.index,df1.values,marker='o')

```

```python
# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3
plt.plot(df1.index,df1.values,marker='o',mfc='red',mec='red',linewidth=3,linestyle='dotted',color='red',label='Total profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
```

```python
# 3. Read all product sales data and show it  using a multiline plot square marker with color=green
month=df.iloc[:,0]
facecream=df.iloc[:,1]
facewash=df.iloc[:,2]
toothpaste=df.iloc[:,3]
bathingsoap=df.iloc[:,4]
shampoo=df.iloc[:,5]
moisturizer=df.iloc[:,6]
lab=['month_number', 'facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer']
plt.plot(month,facecream,marker='s',color='green')
plt.plot(month,facewash,marker='s',color='red')
plt.plot(month,toothpaste,marker='s',color='black')
plt.plot(month,bathingsoap,marker='s',color='blue')
plt.plot(month,shampoo,marker='s',color='cyan')
plt.plot(month,moisturizer,marker='s',color='yellow')
plt.legend(title='all product sales data',labels=lab)




```

```python
# 4.Read toothpaste sales data of each month and show it using a scatterplot
df1=df.groupby('month_number')['toothpaste'].sum()
df1
plt.scatter(df1.index,df1.values)
plt.xticks(df1.index)
plt.xlabel('month_number')
plt.ylabel('toothpaste sales')
plt.show()
```
