```python
# removal of outliers
df['Sale'].unique()
df.drop(['Product_id','Sale'],inplace=True,axis=1)
df
```

```python
def removal_outliers(df,columns,upper_whisker,lower_whisker):
  sns.boxplot(df[columns])
  plt.title(f'Original boxplot with outlier')
  plt.show()
  df=df[df[columns]<=upper_whisker]
  df=df[df[columns]>=lower_whisker]
  sns.boxplot(df[columns])
  plt.title(f'Boxplot after outlier removal')
  plt.show()
  return df

Q1,Q2,Q3=np.percentile(df['weight'],[25,50,75],method='midpoint')
print(Q1,Q2,Q3)
IQR=Q3-Q1
upper_whisker=Q3+1.5*IQR
lower_whisker=Q1-1.5*IQR
print(upper_whisker,lower_whisker)

outlier_free_df=removal_outliers(df,'weight',upper_whisker,lower_whisker)
print(outlier_free_df)

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
for i in df.columns:
  df[i]=scaler.fit_transform(df[[i]])
df
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)

```

```python
from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.svm import SVR

from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,mean_absolute_percentage_error,root_mean_squared_error

lst=[LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(),SVR(),KNeighborsRegressor()]

for i in lst:

  print(f'model is {i}')

  print()

  i.fit(x_train,y_train)

  y_predict=i.predict(x_test)

  r2=r2_score(y_test,y_predict)

  print(f'{i} has R2score {r2}')

  print()

  mape=mean_absolute_percentage_error(y_test,y_predict)

  print(f'{i} has MAPE {mape}')

  print()
```
