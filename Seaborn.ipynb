2.Chi-Square Test:used for extracting relationship of features in object datatype
The Chi-Square test is used to determine if there is a significant association between two categorical variables. In the context of your code and the 'tips' dataset, you might want to use the Chi-Square test to see if there is a relationship between features like 'sex' and 'smoker' or 'day' and 'time'.
Interpreting the Results

Chi-Square Statistic: A measure of the difference between observed and expected frequencies.
P-value: The probability of observing the data if there were no association between the variables. A low p-value (typically less than 0.05) suggests a significant association.
Degrees of Freedom: The number of independent values in the calculation.
Expected Frequencies: The frequencies expected if there were no association between the variables.


```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table (also called a cross-tabulation or crosstab)
# This table shows the frequencies of each combination of categories
contingency_table = pd.crosstab(df['total_bill'], df['tip'])


# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(contingency_table)
# Print the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:", expected)
```

```python
# Boxplot
# outliers can be detected using box plot
sns.boxplot(x='smoker',y='total_bill',data=df)


# The Box

#     The central line: Represents the median of the data. This is the middle value when the data is sorted(50%)
#     The lower edge of the box: Represents the first quartile (Q1), which is the 25th percentile of the data.
# This means that 25% of the data values are below this line.
#     The upper edge of the box: Represents the third quartile (Q3), which is the 75th percentile of the data.
# This means that 75% of the data values are below this line.
#     The height of the box: Represents the interquartile range (IQR),
#  which is the difference between Q3 and Q1 (IQR = Q3 - Q1). This range contains the middle 50% of the data.

# The Whiskers

#     The lines extending from the box (whiskers): Typically represent the range of the data within
#  1.5 times the IQR from the edges of the box.
#         Upper whisker: Extends to the highest data point that is within 1.5 * IQR above Q3.
#         Lower whisker: Extends to the lowest data point that is within 1.5 * IQR below Q1.

# Outliers

#     Points plotted individually beyond the whiskers: These are data points that
# fall outside the typical range and are considered outliers. They can indicate unusual or extreme values in the data
# how to calculate Q2:

# data-7,8,1,1,2,4,3,5
# 1.sort in ascending order 1 1 3 4 5 6 7 8
# 2.divide it in to 2 equal parts   1 1 3 4        5 6 7 8
# Q2=median of whole data (5+4)/2=4.5
# Q3=6.5(median of 2nd part)
# Q1=2(median of first part)
# 3.calculate IQR=Q3-Q1=4.5

# # TOP = Q3+1.5*IQR=6.5+1.5*4.5=13.25              values above 13.25 will be detected as outliers
# BOTTOM=Q1-1.5*IQR=-4.75                           values below 4.75
```

```python
# sns.boxplot(y='total_bill',data=df)
import matplotlib.pyplot as plt
def removal_box_plot(df,columns,threshold):
  sns.boxplot(df[columns])
  plt.title(f'original box plot of {columns}')
  plt.show()
  removed_outliers=df[df[columns]<=threshold]
  print(removed_outliers)
  sns.boxplot(removed_outliers[columns])
  return removed_outliers

threshold=38
df_without_outliers=removal_box_plot(df,'total_bill',threshold)

```
